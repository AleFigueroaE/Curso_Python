Crea una gr치fica que describa cu치ntos alumnos hay por cada grado. La gr치fica debe ser clara y f치cil de interpretar para el usuario. 

```python
import pandas as pd
import numpy as np
import altair as alt

data = pd.read_csv('cleanstudentscomplete.csv')
data.head()
```

```python
grado_count = data['grade'].value_counts().to_frame().reset_index().rename(columns={'index':'grade','grade':'count'}) #Agrupo por grado y hago conteo
grado_count['percent'] = (grado_count['count'] / grado_count['count'].sum()) #Saco el promedio
grado_count
```

```python
bars= alt.Chart(grado_count).mark_bar(stroke = "white").encode(
    x= alt.X("percent:Q", 
        title = "", 
        axis = alt.Axis(format='.0%', grid = False, )),
    y= alt.Y("grade:O", 
        title="",
        axis = alt.Axis(labelAngle=0, labelFontSize=14, labelColor= "#343a40", labelFont="Arial") ),    
    tooltip = [alt.Tooltip('count', title = 'Count' ), alt.Tooltip ('grade:O',title = 'Grade' ) , alt.Tooltip('percent:Q', format='.1%', title = 'Percent' )],
    color=alt.Color(field="grade", type="nominal",legend=None ))

text = bars.mark_text(
    dx=+20,
    size=16,
    fill = "black",
).encode(text= alt.Text('percent:Q', format='.0%'))

(bars + text).properties(
    title = f"Alumnos por Grado",
    width = 600,
    height= 150,
).configure_title(
    fontSize = 20,
    font = "Arial",
).configure_view(
    strokeWidth = 0,)
```

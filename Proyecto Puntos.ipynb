```python
data = pd.read_csv('Puntos.csv',encoding="ISO-8859-1")
data.head()
```

```python
data.rename(columns={"meta de compra":"meta mensual","portafolio diferenciado":"portafolio"},inplace=True)
data.head()
```

```python
tamaño_count = data['tamaño'].value_counts().to_frame().reset_index().rename(columns={'index':'tamaño','tamaño':'count'}) #Agrupo por tamaño y hago conteo
tamaño_count['percent'] = (tamaño_count['count'] / tamaño_count['count'].sum()) #Saco el promedio
tamaño_count
```

```python
bars= alt.Chart(tamaño_count).mark_bar(stroke = "white").encode(
    x= alt.X("percent:Q", 
        title = "", 
        axis = alt.Axis(format='.0%', grid = False, )),
    y= alt.Y("tamaño:O", 
        title="",
        axis = alt.Axis(labelAngle=0, labelFontSize=14, labelColor= "#343a40", labelFont="Arial") ),    
    tooltip = [alt.Tooltip('count', title = 'Count' ), alt.Tooltip ('tamaño:O',title = 'Tamaño' ) , alt.Tooltip('percent:Q', format='.1%', title = 'Percent' )],
    color=alt.Color(field="tamaño", type="nominal",legend=None ))

text = bars.mark_text(
    dx=+20,
    size=16,
    fill = "black",
).encode(text= alt.Text('percent:Q', format='.0%'))

(bars + text).properties(
    title = f"Mix de Tamaño",
    width = 600,
    height= 150,
).configure_title(
    fontSize = 20,
    font = "Arial",
).configure_view(
    strokeWidth = 0,)
```

```python
region_count = data['región'].value_counts().to_frame().reset_index().rename(columns={'index':'región','región':'count'})
region_count['percent'] = (region_count['count'] / region_count['count'].sum())
region_count
```

```python
bars = alt.Chart(region_count).mark_bar(stroke = "white").encode(
    x= alt.X("percent:Q", 
        title = "", 
        axis = alt.Axis(format='.0%', grid = False, )),
    y= alt.Y("región:O", 
        title="",
        axis = alt.Axis(labelAngle=0, labelFontSize=14, labelColor= "#343a40", labelFont="Arial") ),    
    tooltip = [alt.Tooltip('count', title = 'Count' ), alt.Tooltip ('región:O',title = 'Región' ) , alt.Tooltip('percent:Q', format='.1%', title = 'Percent' )],
    color=alt.Color(field="región", type="nominal",legend=None ),
)

text = bars.mark_text(
    dx=+20,
    size=16,
    fill = "black",
).encode(text= alt.Text('percent:Q', format='.0%'))

(bars + text).properties(
    title = f"Distribución por Región",
    width = 700,
    height= 150,
).configure_title(
    fontSize = 20,
    font = "Arial",
).configure_view(
    strokeWidth = 0,
) 
```

```python
tamaño_meta = data.groupby('tamaño')['meta mensual'].mean().to_frame().reset_index() # promedio de cada asignatura segun su tamaño
tamaño_tienda= data.groupby('tamaño')['mi tienda'].mean().to_frame().reset_index()
tamaño_portafolio = data.groupby('tamaño')['portafolio'].mean().to_frame().reset_index()
```

```python
 bars = alt.Chart(tamaño_meta).mark_bar(stroke = "black").encode(
    x= alt.X("tamaño:O", 
        title="",
        axis = alt.Axis(labelAngle=0, labelFontSize=14, labelColor= "#343a40", labelFont="Arial") ),    
    tooltip = [alt.Tooltip('tamaño', title = 'Tamaño' ), alt.Tooltip ('meta mensual:O',title = 'Meta Mensual', format = ".4s" )],
    color=alt.Color(field="tamaño", type="nominal",legend=None ),
    y= alt.Y("meta mensual:Q", 
        title = "Meta Mensual", 
        axis = alt.Axis( grid = False, )),
 )

text = bars.mark_text(
    dy=-10,
    size=14,
    fill = "black",
).encode(text= alt.Text('meta mensual:Q', format = ".3s"))

g_meta = (bars + text).properties(
    width = 300,
)
g_meta
```

```python
 bars = alt.Chart(tamaño_tienda).mark_bar(stroke = "black").encode(
    x= alt.X("tamaño:O", 
        title="",
        axis = alt.Axis(labelAngle=0, labelFontSize=14, labelColor= "#343a40", labelFont="Arial") ),    
    tooltip = [alt.Tooltip('tamaño', title = 'Tamaño' ), alt.Tooltip ('mi tienda:O',title = 'Mi Tienda', format = ".4s" )],
    color=alt.Color(field="tamaño", type="nominal",legend=None ),
    y= alt.Y("mi tienda:Q", 
        title = "Mi Tienda", 
        axis = alt.Axis( grid = False, )),
 )

text = bars.mark_text(
    dy=-10,
    size=14,
    fill = "black",
).encode(text= alt.Text('mi tienda:Q', format = ".3s"))

g_mitienda = (bars + text).properties(
    width = 300,
)
g_mitienda
```

```python
 bars = alt.Chart(tamaño_portafolio).mark_bar(stroke = "black").encode(
    x= alt.X("tamaño:O", 
        title="",
        axis = alt.Axis(labelAngle=0, labelFontSize=14, labelColor= "#343a40", labelFont="Arial") ),    
    tooltip = [alt.Tooltip('tamaño', title = 'Tamaño' ), alt.Tooltip ('portafolio:O',title = 'Portafolio', format = ".4s" )],
    color=alt.Color(field="tamaño", type="nominal",legend=None ),
    y= alt.Y("portafolio:Q", 
        title = "Portafolio", 
        axis = alt.Axis( grid = False, )),
 )

text = bars.mark_text(
    dy=-10,
    size=14,
    fill = "black",
).encode(text= alt.Text('portafolio:Q', format = ".3s"))

g_portafolio = (bars + text).properties(
    width = 300,
)
g_portafolio
```

```python
###Aqui todos juntos
alt.hconcat(g_meta, g_mitienda, g_portafolio ).properties(
    title = 'Performance por Tamaño',
).configure_title(
    fontSize = 18,
    font = "Arial",
    anchor = "middle"
).configure_view(
    strokeWidth = 0,
)
```
